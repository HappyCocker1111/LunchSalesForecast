# データの読み込み


```python
import pandas as pd
sample = pd.read_csv('./sample.csv' , header=None)
sample.head()
```

```python

train = pd.read_csv('./train.csv')
train.head()
```

```python
#説明変数１つの時の予測モデルを作る。soldout。モデルを作るときのデータはtrain。前処理をするときtrainのデータを作る
test = pd.read_csv('./test.csv')
test.head()
```

# 前処理

```python
df=train[['soldout','y']]
```

```python
#trainの前処理をしていくy  soldout
df.info()
```

# モデルの学習

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# 説明変数
X = train[['soldout']]
# 目的変数
y = train[['y']]
```

```python
# 線形回帰モデルのインスタンス生成
model = LinearRegression()

# モデルの学習
model.fit(X, y)
```

```python
# 切片と係数の取得
print(model.intercept_)
print(model.coef_)

# R^2スコアの計算
from sklearn.metrics import r2_score
y_pred = model.predict(X)
print(r2_score(y, y_pred))
```

# 予測

```python
# 新しいデータ
new_X = test[['soldout']]

# 予測される販売数
y_pred = model.predict(new_X)
print(y_pred)
```

# 提出

```python
sample[1] = y_pred
sample
```

```python
# CSVファイルへの書き込み
sample.to_csv("sample.csv", index=False)
```
